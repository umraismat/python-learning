```python
def  new_decorator( original_function):
    def wrap_func():
        print('print some code before  the original function')
        original_function()
        print('print some code after  the  original function')
    return wrap_func
```

```python
def func_need_decorator():
    print('i want to decorate this code')
```

```python
func_need_decorator()
```

```python
decorated_func = new_decorator(func_need_decorator)
```

```python
decorated_func()
```

```python
@new_decorator
def func_need_decorator():
    print('i want to decorate this code')
```

```python
func_need_decorator()
```
