Find PI to the Nth Digit - Enter a number and have the program generate PI up to that many decimal places. Keep a limit to how far the program will go

```python
import math 
# Math library ko laya gaya hai kyunki usme Pi ki value pehle se hoti hai

def easy_pi():
    # User se pucha ja raha hai kitne digits chahiye
    n_input = input("Enter number of decimal places (1-15): ")
    
    # Input ko check kar rahe hain
    try:
        n = int(n_input)
        # Text ko number mein badal rahe hain
        
        # Limit set kar rahe hain (Standard math library 15 digits tak accurate hoti hai)
        if n > 15:
            print("Limit 15 tak hai, kyunki ye basic math version hai!")
        else:
            # Format function use karke Pi ko 'n' places tak dikha rahe hain
            res = format(math.pi, f".{n}f")
            #f".{n}f": A template or mold, waiting to know how many digits to keep. In this example, let’s assume the user entered n = 4.
            print(f"Pi ki value {n} places tak: {res}")
            
    except:
        # Agar user ne galat input diya
        print("Please valid number daaliye.")

easy_pi()
```

 Just like the previous problem, but with e instead of PI. Enter a number and have the program generate e up to that many decimal places. Keep a limit to how far the program will go.

```python
import math # Math library ko laya gaya hai kyunki usme 'e' ki value pehle se hoti hai

def find_e():
    user_input = input("Enter decimal places for 'e' (Max 15): ")
    
    try:
        n = int(user_input) # Input ko text se number mein badal rahe hain
        
        # Limit set kar rahe hain taaki calculation sahi rahe
        if n > 15:
            print("Limit 15 tak hai, kyunki standard math itna hi allow karta hai!")
        elif n < 0:
            print("Bhai, negative number nahi chalega!")
        else:
            #  'e' ko 'n' decimal places tak cut karna
            res = format(math.e, f".{n}f") 
            print(f"Value of 'e' up to {n} places: {res}")
            
    except ValueError:
        print("Please sirf ek number daaliye!")

find_e() # Program ko start kiya ja raha hai
```

```python
def generate_fibonacci():
    
    user_input = input("Kitne Fibonacci numbers chahiye?: ")
    
    try:
        n = int(user_input) 
        a = 0
        b = 1
        
        if n <= 0:
            print("Please 0 se bada number daalein!")
        elif n == 1:
            print(f"Sequence: {a}")
        else:
            print("Sequence:", end=" ")
            # Loop jo n baar chalega
            for i in range(n):
                print(a, end=" ")
                temp = a
                a = b
                b = temp + b
                
    except ValueError:
        print("Oops! Sirf numbers allow hain.")

generate_fibonacci()
```

```python
Prime Factorization - Have the user enter a number and find all Prime Factors (if there are any) and display them
```

```python
def find_prime_factors():
    num = int(input("Enter number: "))
    divisor = 2
    
    print("Prime Factors:", end=" ")
    
    while num > 1:
        if num % divisor == 0:
            print(divisor, end=" ")
            num = num // divisor
        else:
            divisor = divisor + 1
            
    print()

find_prime_factors()
```

```python
def brutal_factor():
    n = int(input("Enter number: "))
    d = 2
    
    print(f"Factors:", end=" ")
    
    while n > 1:
        if n % d == 0:
            print(d, end=" ")
            n = n // d
        else:
            d = d + 1
            
    print("\nDone.")

brutal_factor()
```
