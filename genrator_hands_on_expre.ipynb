```python
def create_cube(n):
    result = []
    for x in range(n):
        result.append(x**3)
    return result    
```

```python
for x in create_cube(10):
    print(x)
```

```python
list(create_cube(10))
```

```python
def gen_fibon(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
for number in gen_fibon(10):
    print(number)
```
