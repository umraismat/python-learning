```python
work_hour = [('inzamam', 4000), ('umra', 200), ('uwaim', 9000), 
             ('Riyan', 6767), ('Ashwin', 6098), ('vIRAT', 90001)]
import random
work_hour = [
    (f"employee_{i}", random.randint(100, 10000)) 
    for i in range(10000)
]

print("=" * 70)
print("ORIGINAL VERSION")
print("=" * 70)

def employee_check_original(GHIJK):
    current_max = 0
    employee_of_year = ''
    promotion = []
    
    for employee, hours in GHIJK:
        promotion.append(hours)
    
    promotion.sort()
    first_promotion_hours = promotion[-1]
    first_promotion_name = ''
    sec_promotion_hours = promotion[-2]
    sec_promotion_name = ''
    
    for employee, hours in GHIJK:
        if hours > current_max:
            current_max = hours
            employee_of_year = employee
        else:
            pass
        
        if hours == first_promotion_hours:
            first_promotion_name = employee
        
        if hours == sec_promotion_hours:
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_original(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 1: Remove unnecessary else and pass statement")
print("=" * 70)

def employee_check_v1(GHIJK):
    current_max = 0
    employee_of_year = ''
    promotion = []
    
    for employee, hours in GHIJK:
        promotion.append(hours)
    
    promotion.sort()
    first_promotion_hours = promotion[-1]
    first_promotion_name = ''
    sec_promotion_hours = promotion[-2]
    sec_promotion_name = ''
    
    for employee, hours in GHIJK:
        if hours > current_max:
            current_max = hours
            employee_of_year = employee
        
        if hours == first_promotion_hours:
            first_promotion_name = employee
        
        if hours == sec_promotion_hours:
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v1(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 2: Use list comprehension for promotion list")
print("=" * 70)

def employee_check_v2(GHIJK):
    current_max = 0
    employee_of_year = ''
    
    # List comprehension instead of loop
    promotion = [hours for employee, hours in GHIJK]
    
    promotion.sort()
    first_promotion_hours = promotion[-1]
    first_promotion_name = ''
    sec_promotion_hours = promotion[-2]
    sec_promotion_name = ''
    
    for employee, hours in GHIJK:
        if hours > current_max:
            current_max = hours
            employee_of_year = employee
        
        if hours == first_promotion_hours:
            first_promotion_name = employee
        
        if hours == sec_promotion_hours:
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v2(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 3: Combine both loops into one")
print("=" * 70)

def employee_check_v3(GHIJK):
    current_max = 0
    employee_of_year = ''
    
    promotion = [hours for employee, hours in GHIJK]
    promotion.sort()
    first_promotion_hours = promotion[-1]
    first_promotion_name = ''
    sec_promotion_hours = promotion[-2]
    sec_promotion_name = ''
    
    # Single loop to find all employees
    for employee, hours in GHIJK:
        if hours > current_max:
            current_max = hours
            employee_of_year = employee
        
        if hours == first_promotion_hours:
            first_promotion_name = employee
        elif hours == sec_promotion_hours:  # Use elif since hours can't match both
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v3(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 4: Use max() function instead of manual tracking")
print("=" * 70)

def employee_check_v4(GHIJK):
    # Find employee with maximum hours using max()
    employee_of_year, current_max = max(GHIJK, key=lambda x: x[1])
    
    promotion = [hours for employee, hours in GHIJK]
    promotion.sort()
    first_promotion_hours = promotion[-1]
    first_promotion_name = ''
    sec_promotion_hours = promotion[-2]
    sec_promotion_name = ''
    
    for employee, hours in GHIJK:
        if hours == first_promotion_hours:
            first_promotion_name = employee
        elif hours == sec_promotion_hours:
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v4(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 5: Use sorted() instead of sort() to avoid mutation")
print("=" * 70)

def employee_check_v5(GHIJK):
    employee_of_year, current_max = max(GHIJK, key=lambda x: x[1])
    
    # Use sorted() to get sorted list without mutating
    sorted_hours = sorted([hours for employee, hours in GHIJK])
    first_promotion_hours = sorted_hours[-1]
    first_promotion_name = ''
    sec_promotion_hours = sorted_hours[-2]
    sec_promotion_name = ''
    
    for employee, hours in GHIJK:
        if hours == first_promotion_hours:
            first_promotion_name = employee
        elif hours == sec_promotion_hours:
            sec_promotion_name = employee
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v5(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 6: Sort tuples directly by hours (descending)")
print("=" * 70)

def employee_check_v6(GHIJK):
    # Sort all employees by hours in descending order
    sorted_employees = sorted(GHIJK, key=lambda x: x[1], reverse=True)
    
    # First employee has max hours
    employee_of_year, current_max = sorted_employees[0]
    first_promotion_name = sorted_employees[0][0]
    sec_promotion_name = sorted_employees[1][0]
    
    return (employee_of_year, current_max, first_promotion_name, sec_promotion_name)

# result = employee_check_v6(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 7: Use unpacking for cleaner code")
print("=" * 70)

def employee_check_v7(GHIJK):
    sorted_employees = sorted(GHIJK, key=lambda x: x[1], reverse=True)
    
    # Unpack first two employees
    (first_name, first_hours), (second_name, second_hours) = sorted_employees[:2]
    
    return (first_name, first_hours, first_name, second_name)

# result = employee_check_v7(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 8: Most Pythonic - Using nlargest for efficiency")
print("=" * 70)

from heapq import nlargest

def employee_check_v8(GHIJK):
    # Get top 2 employees by hours - O(n log k) where k=2
    top_two = nlargest(2, GHIJK, key=lambda x: x[1])
    
    first_name, first_hours = top_two[0]
    second_name, _ = top_two[1]
    
    return (first_name, first_hours, first_name, second_name)

# result = employee_check_v8(work_hour)
# print(f"Result: {result}\n")


print("=" * 70)
print("VERSION 9: Handle edge cases (empty list, single employee)")
print("=" * 70)

from heapq import nlargest

def employee_check_v9(GHIJK):
    if not GHIJK:
        return ('', 0, '', '')
    
    if len(GHIJK) == 1:
        name, hours = GHIJK[0]
        return (name, hours, name, '')
    
    top_two = nlargest(2, GHIJK, key=lambda x: x[1])
    first_name, first_hours = top_two[0]
    second_name, _ = top_two[1]
    
    return (first_name, first_hours, first_name, second_name)

# result = employee_check_v9(work_hour)
# print(f"Result: {result}\n")

# # Test edge cases
# print("Edge case - Empty list:")
# print(employee_check_v9([]))

# print("\nEdge case - Single employee:")
# print(employee_check_v9([('John', 5000)]))


print("\n" + "=" * 70)
print("PERFORMANCE COMPARISON")
print("=" * 70)

import time

def benchmark(func, data, iterations=10000):
    start = time.time()
    for _ in range(iterations):
        func(data)
    end = time.time()
    return end - start

print(f"Original version: {benchmark(employee_check_original, work_hour):.4f}s")
print(f"Version1: {benchmark(employee_check_v1, work_hour):.4f}s")
print(f"Version 2: {benchmark(employee_check_v2, work_hour):.4f}s")
print(f"Version 3: {benchmark(employee_check_v3, work_hour):.4f}s")
print(f"Version 4: {benchmark(employee_check_v4, work_hour):.4f}s")
print(f"Version 5: {benchmark(employee_check_v5, work_hour):.4f}s")
print(f"Version 6: {benchmark(employee_check_v6, work_hour):.4f}s")
print(f"Version 7: {benchmark(employee_check_v7, work_hour):.4f}s")
print(f"Version 8 (nlargest): {benchmark(employee_check_v8, work_hour):.4f}s")
```
